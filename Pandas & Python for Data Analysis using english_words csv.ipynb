create a column Ratio which represents the 'Freq Ratio' of a word

```python
df['Char Count']= df.index.str.len()
df
```

```python
df['Ratio'] = df['freq']/df['Char Count']
df.sample(50)
```

what is the maximum value of ratio?

```python
df['Ratio'].describe()
```

what word is the one with the highest ratio?

```python
df.sort_values(by='Ratio',ascending=False).head()
```

```python
df.loc[df['Ratio'] == df['Ratio'].max()]
```

What is the shortest wrod of freq = 274?

```python
df.loc[
    (df['freq'] == 274) &
    (df['Char Count'] == df.loc[df['freq'] == 274, 'Char Count'].min())
]
```
